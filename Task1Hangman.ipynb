```python
import random
Name=input("Enter your name:")
print(f"Good luck!{Name}")

hangman_stages = [
    """
      |------------
      |       |
      |
      |
      |
    __|_________
    """,
    """
      |------------
      |       |
      |       O
      |
      |
    __|_________
    """,
    """
      |------------
      |       |
      |       O
      |       |
      |
    __|_________
    """,
    """
      |------------
      |       |
      |       O
      |      \|
      |
    __|_________
    """,
    """
      |------------
      |       |
      |       O
      |      \|/
      |
    __|_________
    """,
    """
      |------------
      |       |
      |       O
      |      \|/
      |      /
    __|_________
    """,
    """
      |------------
      |       |
      |       O
      |      \|/
      |      / \
    __|_________
    """
]

words_with_clues={'java':'Programming language',
                  'python':'Programming language',
                  'sql':'a standardized programming language that is used to manage relational databases ',
                  'artificial':'the simulation of human intelligence processes by machines',
                  'mouse':'cursor',
                  'computer':'an electronic device that manipulates information, or data'}
word,clue=random.choice(list(words_with_clues.items()))
word_dis=["_"]*len(word)
guesses=[]
lives=len(hangman_stages)-1
print(f"Your Clue is {clue}")
while lives>0 and "_" in word_dis:
    print(f"Hey {Name}! {lives} Lives are left")
    print(" ".join(word_dis))
    print(hangman_stages[len(hangman_stages)-lives-1])

    Guess=input("Enter you Guessed Character:")
    if Guess in guesses:
        print(f"you are already guessed the character {Guess}! and Try to Guess Another letter")
    elif Guess in word:
        print(f"Your Guess is Correct,{Guess} is in the word")
        for i in range(len(word)):
            if word[i]==Guess:
                word_dis[i]=Guess
    else:
        print("Sorry your Guess is Wrong,Try Again")
        lives-=1
    guesses.append(Guess)

if "_" not in word_dis:
    print(f"Congratulation {Name} Your all guesses is Correct")
else:
    print(f"Sorry You lose the game better luck next time,The Correct Answer is {word}")


```
